# Which country has how many percent people

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dataFrame = pd.read_csv('new.csv')
```

```python
data = dataFrame['Country'].value_counts()
```

```python
data
```

```python
data.head(20)
```

```python
df = data.head(20)
```

```python
df.plot.pie(y=df.index,
           shadow=False,
           startangle=90,
           figsize=[10, 10],
           autopct='%1.1f%%')

plt.axis('equal')
plt.tight_layout()
plt.show()
```
